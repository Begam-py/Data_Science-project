# Select the top 10 countries based on 1960 to 2023

```python
years = [str(year) for year in range(1960, 2024)]  # List of years from 1960 to 2023
df['Total Population'] = df[years].sum(axis=1)  # Sum population data for each country
df_sorted = df.sort_values(by='Total Population', ascending=False)
top_10_countries = df_sorted.head(10)
print(top_10_countries[['Country Name', 'Total Population']])
```

```python
colors = np.random.rand(10, 4)
plt.figure(figsize=(10, 6))
plt.bar(top_10_countries['Country Code'], top_10_countries['Total Population'],color=colors)
plt.xlabel('Country Code')
plt.ylabel('Total Population')
plt.title('Top 10 Countries by Total Population (1960-2023)')
plt.tight_layout()
plt.show()
```
